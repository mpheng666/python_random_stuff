# What to do?  CLEANING!!!

#### Here is the possible code that can help us to clean

```python
df.dropna(inplace=True)
```

```python
import numpy as np

for i in df.columns:
#replacing the Title name with empty string
    df[i]=df[i].str.replace(i,'').astype(str)
#if comma is found, replace with empty string
    df[i]=df[i].str.replace(",",'').astype(str)
#if not Borough, convert to integer
    if i!='Borough':
        df[i] = pd.to_numeric(df[i], errors='coerce', downcast='integer')
```
