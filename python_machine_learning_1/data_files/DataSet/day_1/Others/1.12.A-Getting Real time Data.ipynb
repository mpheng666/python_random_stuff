```python
import bs4          
import requests     
import pandas as pd
```

```python
URL = 'https://sg.finance.yahoo.com/currencies'
```

```python
response = requests.get(URL)
```

```python
response.status_code
```

```python
len(response.content)
```

```python
beausoup=bs4.BeautifulSoup(response.text, 'lxml') 
```

```python
html.parser
```

```python
beausoup
```

```python
myresult=[]
```

```python
table = beausoup.find("table") # Locate the "table" tag 
```

```python
rows=table.find_all("tr")# Find the row within table
```

```python
len(rows)
```

```python
type(rows)
```

```python
rows
```

```python
myresult=[]
for row in rows:
    cells=row.find_all("td")# Find cell
    cells = cells[1:3]
    for cell in cells:
        #print(cell.text)
        myresult.append([cell.text for cell in cells])
```

```python
myresult
```

```python
pd.DataFrame(myresult, columns=["currency", "rate"]).drop(0, axis=0)
```
