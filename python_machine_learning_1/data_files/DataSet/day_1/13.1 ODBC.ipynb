# 13.1. Open Database Connectivity (ODBC)

## This example must be done on the server (172.29.180.96:1000x)

ODBC (Open Database Connectivity) is a technology that allows one to connect to a whole variety of databases using a single API, the ODBC API.

![ODBC](fig/ODBC.png)

ODBC driver manager provides  ways to make a shortcuts (DNS) for a set of parameters:

DSN (DataSource Name): a system or user defined data source, identified by a (unique) name. DSNs can be configured using the ODBC Data Source Administrator).
File DSN: a file which contains connection parameters. An ODBC Data Source Administrator usually allows you to create File DSNs

![ODBC](fig/ODBC1.png)
