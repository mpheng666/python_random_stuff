# 2.10 Boxplot

According to Wikipedia (https://en.wikipedia.org/wiki/Box_plot), A box plot or boxplot can be used to describe groups of numerical data through their quartiles. It has the following charateristics:
-  It is one of the most common type of plot. 
-  It summarises one or several numeric variables. 
-  The line that divides the box into 2 parts is <b> the median </b>.
-  The end of the box shows the upper and lower quartiles. 
-  The extreme lines shows the highest and lowest value 
-  The dots outside the extreme lines are outliers.

## 2.10.2 Misleading Boxplots

When boxplots summarizes the data, it compresses the information so that some meaningful information is hidden.

When we look at the below boxplot, we consider group C has a higher value the rest of the variables.  <b>Because we could not look at the detailed information of the variable, we cannot see the actual distribution.
![Good Hdistogram](resources\img\misleading_boxplot.png)


<b>When we see the actual distribution below, we can see that group C only has limited observed data and the average is lower than other variables.
![Good Hdistogram](resources\img\correct_boxplot.png)

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
 
# Dataset:
a = pd.DataFrame({ 'group' : np.repeat('A',5000), 'value': np.random.normal(10, 5, 5000) })
b = pd.DataFrame({ 'group' : np.repeat('B',5000), 'value': np.random.normal(13, 1.2, 5000) })
c = pd.DataFrame({ 'group' : np.repeat('B',5000), 'value': np.random.normal(18, 1.2, 5000) })
d = pd.DataFrame({ 'group' : np.repeat('C',200), 'value': np.random.normal(25, 4, 200) })
e = pd.DataFrame({ 'group' : np.repeat('D',1000), 'value': np.random.uniform(12, size=1000) })
df=a.append(b).append(c).append(d).append(e)
plt.subplots(figsize=(23, 10))
# Usual boxplot
sns.boxplot(x='group', y='value', data=df)
```

```python
plt.subplots(figsize=(23, 10))
ax = sns.boxplot(x='group', y='value', data=df)
ax = sns.stripplot(x='group', y='value', data=df, color="orange", jitter=0.2, size=2.5)

plt.title("Boxplot with jitter", loc="left")
```

## 2.10.3. Boxplot using Matplotlib

```python
import pandas as pd
import numpy as np
```

```python
a = pd.DataFrame({ 'group' : np.repeat('A',5000), 'value': np.random.normal(10, 5, 5000) })
b = pd.DataFrame({ 'group' : np.repeat('B',5000), 'value': np.random.normal(13, 1.2, 5000) })
c = pd.DataFrame({ 'group' : np.repeat('C',5000), 'value': np.random.normal(18, 1.2, 5000) })
d = pd.DataFrame({ 'group' : np.repeat('D',5000), 'value': np.random.normal(25, 4, 5000) })
e = pd.DataFrame({ 'group' : np.repeat('E',5000), 'value': np.random.uniform(12, size=5000) })
```

```python
frames = [a, b, c, d, e]
df1=pd.concat(frames)
df2=pd.concat(frames, axis=1)
a
```

##### Preparing format for SNS

```python
df1.sample(2) #tidy format for sns
df1
```

##### Preparing format for Matplotlib and Pandas.Plot

```python
df2.sample(2)
df2
```

```python
df2.drop(['group'], axis=1, inplace=True)
```

```python
df2.columns=['A', 'B', 'C', 'D', 'E']
```

```python
df2.sample(2)
df2
```

```python

import matplotlib.pyplot as plt
#from matplotlib import rcParams
# figure size in inches
plt.rcParams['figure.figsize'] = 6.5,4
#plt.rcdefaults()

```

```python
plt.boxplot(df2.values, labels=df2.columns)
plt.show()
```

# Examples using Cars DataSet

## Using Pandas.Plot with Cars Data

```python
import pandas as pd
# load cars.csv data into dataframe named cars
cars = pd.read_csv("cars.csv",index_col=None)
```

```python
cars.sample(2)
```

```python
df=cars[['Cylinders','MPG']]
```

```python
df.boxplot(column='MPG', by='Cylinders')
```

## Using Seaborn with Cars Data

```python
import seaborn as sns
sns.boxplot(y="MPG", x="Cylinders", data=df)
```
