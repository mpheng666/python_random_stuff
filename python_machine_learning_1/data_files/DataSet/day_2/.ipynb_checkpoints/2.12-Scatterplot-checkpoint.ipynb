# 2.12 Scatterplot
A scatterplot is a plot with two-value axes, showing a set of numerical data along a vertical axis and another set of numerical values along a horizontal axis. The plot shows points at the intersection of an x and y numerical value, combining the two values into single data points. Depending on the data, the points may be distributed evenly or unevenly in the horizontal axis. Scatter plots are usually used for:
- understanding the distribution of the data
- showing relationships between 2 (scatter) or more numerical values (usually differentiated using colors).
- highlighting linear or non-linear trends, clusters, and outliers.
- creating two groups of numbers as one series of xy coordinates.



## Guideline for Line Charts:
- Avoid analyzing too many categories (at the most 3)
- Include more data in a scatter chart so the comparisons that we make are better.
- Begin they-axis at 0 to represent data accurately.
- To make the plot easier to understand, only use a maximum of two variables if trend lines are plotted, .


## 2.11.1. Scatter Plot using Matplotlib

```python
import matplotlib.pyplot as plt
plt.rcdefaults()
plt.rcParams['figure.figsize']=7,4
from numpy.random import random
```

```python
colors = ['b', 'c', 'y', 'm', 'r']

E = plt.scatter(random(10), random(10), marker='x', color=colors[0])
W = plt.scatter(random(10), random(10), marker='o', color=colors[0])
S  = plt.scatter(random(10), random(10), marker='o', color=colors[1])
N = plt.scatter(random(10), random(10), marker='o', color=colors[2])
NE  = plt.scatter(random(10), random(10), marker='o', color=colors[3])
NW = plt.scatter(random(10), random(10), marker='o', color=colors[4])
SE = plt.scatter(random(10), random(10), marker='x', color=colors[4])

plt.legend((E, W, S, N, NE, NW, SE),
           ('East', 'West', 'South', 'North', 'NorthEast', 'NorthWest', 'SouthEast'),
           scatterpoints=1,
           loc='lower left',
           ncol=3,
           fontsize=8)

plt.show()
```

##### Creating Pandas DataFrames

```python
#Creating the same dataframe
Edf=pd.DataFrame({'x':E.get_offsets()[:,0], 'y':E.get_offsets()[:,1]})

Wdf=pd.DataFrame({'x':W.get_offsets()[:,0], 'y':W.get_offsets()[:,1]})

Sdf=pd.DataFrame({'x':S.get_offsets()[:,0], 'S':W.get_offsets()[:,1]})

Ndf=pd.DataFrame({'x':N.get_offsets()[:,0], 'S':N.get_offsets()[:,1]})
            
NEdf=pd.DataFrame({'x':NE.get_offsets()[:,0], 'S':NE.get_offsets()[:,1]})
                     
NWdf=pd.DataFrame({'x':NW.get_offsets()[:,0], 'S':NW.get_offsets()[:,1]})
              
SEdf=pd.DataFrame({'x':SE.get_offsets()[:,0], 'S':SE.get_offsets()[:,1]})
SEdf.sample(3)
```

#### Plotting the Data to Scatter Plot

```python
ax = Edf.plot(kind='scatter', x="x", y="y", color='b', label='East',fontsize=8, marker='x');

### SEE the  ax=ax..... configuring in the same figure ax
Wdf.plot(kind='scatter', x=0, y=1, color='b', label='West',fontsize=8, marker='o', ax=ax)
Sdf.plot(kind='scatter', x=0, y=1, color='c', label='South',fontsize=8, marker='o', ax=ax)
Ndf.plot(kind='scatter', x=0, y=1, color='y', label='North',fontsize=8, marker='o', ax=ax)
NEdf.plot(kind='scatter', x=0, y=1, color='m', label='NorthEast',fontsize=8, marker='o', ax=ax)
NWdf.plot(kind='scatter', x=0, y=1, color='r', label='NorthWest',fontsize=8, marker='o', ax=ax)
SEdf.plot(kind='scatter', x=0, y=1, color='r', label='SouthEast',fontsize=8, marker='x', ax=ax)

plt.legend(ncol=3, loc='lower left', scatterpoints=1, fontsize=8)
plt.show()

```

## 2.11.3. Scatter Plot using Seaborn

```python
import pandas as pd
import numpy as np
df=pd.DataFrame({"X":np.random.random_sample(100), 
              "y":np.random.random_sample(100), 
              "group": ['A'] * 20 + ['B'] * 20 + ['C'] * 20 + ['D'] * 20 + ['E'] * 20 })
```

```python
df.sample(3)
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x="X", y="y", hue="group", data=df)
plt.show()
```

# Examples using Cars DataSet

```python
import pandas as pd
# load cars.csv data into dataframe named cars
cars = pd.read_csv("cars.csv",index_col=None)
```

```python
cars.sample(3)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
countries=cars.groupby('Origin')
```

```python
#fig, ax =plt.subplots()
plt.rc('figure', figsize=(10, 6))
for name, country in countries:
    plt.scatter(country.Horsepower, country.MPG, marker='o', linewidths=8, alpha=.5, label=name)
```

```python
import pandas as pd
# load cars.csv data into dataframe named cars
cars = pd.read_csv("cars.csv",index_col=None)
```

```python
df1=cars.loc[(cars==0).any(axis=1)==False]
```

```python
countries1=df1.groupby('Origin')
plt.rc('figure', figsize=(10, 6))
for name1, country1 in countries1:
    plt.scatter(country1.Horsepower, country1.MPG, marker='o', linewidths=8, alpha=.5, label=name1)
```

```python
df2=df1[['Horsepower', 'MPG', 'Origin']]
```
