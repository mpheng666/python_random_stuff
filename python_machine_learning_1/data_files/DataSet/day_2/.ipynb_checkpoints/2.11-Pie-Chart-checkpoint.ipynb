# 2.11 Pie Chart

A pie chart shows the composition of categories,k illustrating a static number by showing how categories represent part of a whole. A pie chart represents numbers in percentages, and the total sum of all segments needs to equal 100%.  Apply a pie chart to show how individual parts make up the whole of something.  For instance, meat consumption by countries, device types that web visitors used.

## Guideline for Line Charts:
- Avoid analyzing too many categories (more than 6 is too many)
- Include labels to make sure the differentiation between slices.
- Check that the slice values must be add up to 100%.
- Order slices according to their size.

### Label should be included to describe the percentage clearly
Without labels, a pie with similar numbers cannot be detected.  Below is a pie chart <b> without a label </b> :

# 2.10.1. Pie Chart Using Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
 
# Data to plot
labels = ['Singapore', 'Indonesia', 'Malaysia', 'Vietnam']
sizes = [315, 230, 345, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True, startangle=140)
#plt.pie(sizes, explode=explode, labels=labels, colors=colors, shadow=True, startangle=10)

plt.axis('equal')
plt.title('Meat Consumption')
plt.show()
```

# Examples using Cars Dataset

## using MatlplotLib with Cars Data

```python
import pandas as pd
# load cars.csv data into dataframe named cars
cars = pd.read_csv("cars.csv",index_col=None)
```

```python
df=cars[['MPG', 'Origin']]
```

```python
dfmean=df.groupby('Origin').mean()
```

```python
df_new=dfmean.reset_index()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
 
# Data to plot
labels = df_new['Origin'].values
sizes = df_new['MPG'].values
colors = ['gold', 'yellowgreen', 'lightcoral']
explode = (0.1, 0, 0, )  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True, startangle=140)
#plt.pie(sizes, explode=explode, labels=labels, colors=colors, shadow=True, startangle=10)

plt.axis('equal')
plt.title('MPG by Origin')
plt.show()
```

```python
dfmean.plot.pie(y="MPG", figsize=(5, 5), autopct='%1.1f%%', colors=colors,
                   pctdistance=.5, labeldistance=1.2, explode=explode, startangle=140)
plt.show()
```
