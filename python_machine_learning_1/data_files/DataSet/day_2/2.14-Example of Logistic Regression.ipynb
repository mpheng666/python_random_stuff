##### Assigning the target_names to class_names

````python
class_names=iris.target_names
````

##  5 Steps of Applying Scikit-Learn estimator API

 1. Choose Model
 2. Choose model hyperparameters
 3. Arrange Data into a features matrix and target vector
 4. Fit the model
 5. Apply the model



### - 1st step: Choose Model
from sklearn.linear_model import LogisticRegression

````python
from sklearn.linear_model import LogisticRegression
````

Creating an instance of Logistic Regression Algorithm (Initialising Model)

### - 2nd step: Choose model hyperparameters

````python
logClassifier = LogisticRegression()
#using default parameters 
````

#### Confirm the hyperparameter

````python
logClassifier
````

### - 3rd Step: Arrange Data into a Features Matrix and Target Vector

````python
X=iris.data

y=iris.target
X.shape
````

#### Train_Test_Split

````python
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.10, 
                                                    random_state=0)
````

### -4th Step: Fit the model to your data

Put the instance of the logistic regression classifier (i.e., *logClassifier*), calling the *fit* method to ***train*** the model with the training dataset

````python

logClassifier.fit(X_train, y_train)
````

### -5th Step: Predict labels for unknown (unseen) data

Feeding the test dataset to the model and get the results

````python
y_pred = logClassifier.predict(X_test)
````

````python
y_pred
````

##### The Actual Label

````python
y_test
````

````python
metrics.accuracy_score(y_test, y_pred) 
````

##### Metric Performance

````python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred, target_names=class_names))
print(classification_report(y_test, y_pred))
````

### Five-fold cross-validation:

````python
from sklearn.model_selection import cross_val_score
````

````python
logClassifier1 = LogisticRegression()
````

````python
scores = cross_val_score(logClassifier1, iris.data, iris.target, cv=10)
````

````python
scores
````

````python
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
````

### LeaveOneOut

````python
from sklearn.model_selection import LeaveOneOut
````

````python
scoresLOO = cross_val_score(logClassifier1, iris.data, iris.target, cv=LeaveOneOut())
````

````python
scoresLOO.mean()
````
