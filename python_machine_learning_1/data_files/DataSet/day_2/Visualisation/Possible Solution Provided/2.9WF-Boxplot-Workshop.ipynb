```python
import pandas as pd
df = pd.read_csv("resale-sample-regression.csv", index_col=0)
labels=df.Region.unique()
x_pos=labels
labels
```

```python
import pandas as pd
import matplotlib.pyplot as plt
#USING SEABORN

import seaborn as sns
```

```python
data = pd.concat([df['resale_price'], df['Region']], axis=1)
data.head(3)
```

```python

sns.set(font_scale=5) 
#data = pd.concat([df['resale_price'], df['Region']], axis=1)
f, ax = plt.subplots(figsize=(35, 15))
fig = sns.boxplot(x="Region", y="resale_price", data=data)
fig.axis(ymin=5, ymax=1300000);
```
