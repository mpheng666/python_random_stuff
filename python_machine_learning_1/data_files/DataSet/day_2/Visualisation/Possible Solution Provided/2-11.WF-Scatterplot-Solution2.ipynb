# 2-11.WF-Scatterplot

```python
import pandas as pd
%matplotlib inline
df = pd.read_csv("resale-sample-regression.csv", index_col=0)
labels=df.Region.unique()
x_pos=labels
```

```python
df1=df.set_index(['Region', 'floor_area_sqm'])
```

```python
df1.head(3)
```

```python
df1.sort_index(level=['Region','floor_area_sqm'], inplace=True)
```

```python
df1.head(5)
```

```python
df1.xs('Central')['resale_price'].head()
```

```python
df1.xs('Central')['resale_price'].index[0:4]
```

```python
import matplotlib.pyplot as plt
from numpy.random import random
plt.rc('figure', figsize=(15, 5))

colors = ['b', 'c', 'y', 'm']

NEg = plt.scatter(df1.xs('North East')['resale_price'], df1.xs('North East')['resale_price'].index, marker='x', color=colors[0])
Eg = plt.scatter(df1.xs('East')['resale_price'], df1.xs('East')['resale_price'].index, marker='o', color=colors[0])
Cg  = plt.scatter(df1.xs('Central')['resale_price'], df1.xs('Central')['resale_price'].index, marker='o', color=colors[1])
Wg = plt.scatter(df1.xs('West')['resale_price'], df1.xs('West')['resale_price'].index, marker='o', color=colors[2])
Ng  = plt.scatter(df1.xs('North')['resale_price'], df1.xs('North')['resale_price'].index, marker='o', color=colors[3])

plt.legend((NEg,Eg,Cg,Wg,Ng),(labels), scatterpoints=1, loc='lower left', ncol=5, fontsize=8)


plt.xlabel('Resale Price')
plt.ylabel('Floor size (sqm)')

plt.title('Resale Price and Floor size by Region')

plt.show()
```
