# 2-11.S.WF-Scatterplot-Seaborn


```python
import pandas as pd
%matplotlib inline
df = pd.read_csv("resale-sample-regression.csv", index_col=0)
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.rc('figure', figsize=(19, 9))
```

```python
sns.scatterplot(x="floor_area_sqm", y="resale_price", hue="Region",  data=df)
plt.show()
```

```python
sns.scatterplot(x="floor_area_sqm", y="resale_price", hue="Region", size="resale_price",  data=df)
plt.show()
```

```python
sns.scatterplot(x="floor_area_sqm", y="resale_price", hue="Region", size="resale_price", sizes=(30,300),  data=df)
plt.show()
```

```python
sns.scatterplot(x="floor_area_sqm", y="resale_price", hue="Region", size="resale_price", sizes=(30,300),  data=df, style="Region")
plt.show()
```
