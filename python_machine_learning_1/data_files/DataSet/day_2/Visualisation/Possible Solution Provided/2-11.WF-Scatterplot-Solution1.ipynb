```python
import pandas as pd
df = pd.read_csv("resale-sample-regression.csv", index_col=0)
labels=df.Region.unique()
x_pos=labels
```

```python
#Getting Floor Size by Region
df[df['Region']=="North East"]['resale_price'].head(3)
```

```python
#Count how many transaction (resale price) happened in North East
df[df['Region']=="North East"].count()
```

```python
#Assign transactions (resale price) in different region (NE, E, C, etc)
NE=df[df['Region']=="North East"]['resale_price']
E=df[df['Region']=="East"]['resale_price']
C=df[df['Region']=="Central"]['resale_price']
W=df[df['Region']=="West"]['resale_price']
N=df[df['Region']=="North"]['resale_price']
```

```python
#show different type of floor size in NE
df[df['Region']=="North East"]['floor_area_sqm'].head(3)
```

```python
#Count how many different type of floor size in North East
df[df['Region']=="North East"]['floor_area_sqm'].unique()
```

```python
NE_size=df[df['Region']=="North East"]['floor_area_sqm']
E_size=df[df['Region']=="East"]['floor_area_sqm']
C_size=df[df['Region']=="Central"]['floor_area_sqm']
W_size=df[df['Region']=="West"]['floor_area_sqm']
N_size=df[df['Region']=="North"]['floor_area_sqm']
```

```python
import matplotlib.pyplot as plt
from numpy.random import random
plt.rc('figure', figsize=(10, 5))

colors = ['b', 'c', 'y', 'm']

NEg = plt.scatter(NE, NE_size, marker='x', color=colors[0])
Eg = plt.scatter(E, E_size, marker='o', color=colors[0])
Cg  = plt.scatter(C, C_size, marker='o', color=colors[1])
Wg = plt.scatter(W, W_size, marker='o', color=colors[2])
Ng  = plt.scatter(N, N_size, marker='o', color=colors[3])

plt.legend((NEg,Eg,Cg,Wg,Ng),(labels), scatterpoints=1, loc='lower left', ncol=5, fontsize=8)


plt.xlabel('Resale Price')
plt.ylabel('Floor size (sqm)')

plt.title('Resale Price and Floor size by Region')

plt.show()
```
