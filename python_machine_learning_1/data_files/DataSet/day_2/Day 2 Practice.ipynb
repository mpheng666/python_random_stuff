# Practice 2: Data files

```python
%matplotlib inline
import matplotlib.pyplot as plt; 
plt.rcdefaults()
import numpy as np
from matplotlib import rcParams
import pandas as pd
```

Open and read the contents of the resale.csv data file.

```python
df = pd.read_csv("resale.csv")
```

Find and print the most expensive flat

```python
print("Most expensive flat is", (df['price']).max(), "at index", 
      df['price'].idxmax())
```

Find and print the cheapest flat

```python
print("Cheapest flat is", (df['price']).min(), "at index", 
      df['price'].idxmin())
```

Find and print the average and variance of the flat’s prices. 

```python
print("The average of flat's price is :", 
      "{:.2f}".format(df['price'].mean()))

print("The variance of flat's price is :", 
      "{:.2f}".format(df['price'].var()))
```

How many flats are in the WOODLANDS?

```python
print("There are", df['price'].count(), "flats at WOODLANDS. ")
```
