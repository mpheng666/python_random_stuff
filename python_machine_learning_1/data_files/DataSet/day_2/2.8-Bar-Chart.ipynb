# 2.8 Bar Chart

<H2>Guideline for  Bar Charts:</H2>
- Consistent colors should be used
- Choose colors only when we need to highlight meaningful data points or changes over time.
- Include labels to improve readability (horizontal and vertical).
- Start the y-axis at 0 to appropriately reflect the values in your graph.

##  2.8.1. Horizontal Bar vs Vertical Bar

Horizontal Bar is useful if you have many categories (e.g., more than 9) to compare.  The example is as below:

- It is easier to see the differences between Perl and C# in a vertical bar chart
- The labels within a horizontal bar chart are not affected by the lengths of the words (e.g., Cobol and Pascal).

# Matplotlib 

## 2.8.2. A Horizontal Chart Using Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt; 
plt.rcdefaults()
import numpy as np
from matplotlib import rcParams
import pandas as pd
# figure size in inches
rcParams['figure.figsize'] = 8,3
 
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp', 'Basic', 'C#', 'PhP', 'Cobol', 'Pascal')
x_pos = np.arange(len(objects))
print(type(x_pos))
print(x_pos)
performance = [10,8,6,4.5,2,1,6,5,4,5,10]
 
    
N = 13
df = pd.Series(np.random.randint(10,50,N), index=np.arange(1,N+1))

cmap = plt.cm.tab10

colors = cmap(np.arange(len(df)) % cmap.N)
plt.barh(x_pos, performance, align='center', alpha=1, color=colors)
plt.yticks(x_pos, objects)
plt.xlabel('Usage')
plt.title('Programming language usage')
 
plt.show()
```

## 2.8.3. A Vertical Chart Using Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt; 
plt.rcdefaults()
import numpy as np
 
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp', 'Basic', 'C#', 'PhP', 'Cobol', 'Pascal')
y_pos = np.arange(len(objects))
performance = [10,8,6,4.5,2,1,6,5,4,.5, 1]
 
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.title('Programming language usage')
 
plt.show()
```

# Example using Cars Dataset

## Using Matplotlib with Cars Data

```python
import pandas as pd
# load cars.csv data into dataframe named cars
cars = pd.read_csv("cars.csv",index_col=None)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cars.sample(3)
# cars.Origin
```

```python
groups=cars.groupby([cars.Origin])
print(groups)
```

```python
for name, group in groups:
    #print (name)
    #print(group.MPG.mean())
    plt.bar(name, group.MPG.mean(), align='center', alpha=0.5)
    plt.text(name, group.MPG.mean(), "{:.2f}".format(group.MPG.mean()), va='center' )

plt.ylabel('MPG')
plt.title('The Average Miles per Gallon by Country Origin')
plt.show()    
    
    #ax.plot(group.floor_area_sqm, group.resale_price, marker='o', linestyle='', ms=10, label=name)
#ax.legend(numpoints=1)
```

```python
for name, group in groups:
    #print (name)
    #print(group.MPG.mean())
    plt.barh(name, group.MPG.mean(), align='center', alpha=0.5)
    plt.text(group.MPG.mean(), name, "{:.2f}".format(group.MPG.mean()), va='baseline' )

plt.xlabel('MPG')
plt.title('The Average Miles per Gallon by Country Origin')
plt.show() 
```

```python
gr2 = cars.groupby([cars.Cylinders])
# cars.Cylinders

for name, group in gr2:
    #print (name)
    #print(group.MPG.mean())
#     plt.bar(name, group.MPG.mean(), align='center', alpha=0.5)
    plt.bar(name, group.MPG.mean(), align='center', alpha=0.5)
#     plt.text(name, group.MPG.mean(), "{:.2f}".format(group.MPG.mean()), va='center' )
    plt.text(name, group.MPG.mean(), "{:.2f}".format(group.MPG.mean()), va='center' )

plt.ylabel('MPG')
plt.title('The Average Miles per Gallon by Country Origin')
plt.show()    
```

## Using Pandas.plot with Cars Data

```python
import pandas as pd
# load cars.csv data into dataframe named cars
cars = pd.read_csv("cars.csv",index_col=None)
```

```python
groups=cars.groupby([cars.Origin])
```

```python
df=pd.DataFrame(groups.MPG.mean())
```

```python
df
```

```python
import numpy as np
ax = df.plot.bar(y='MPG', rot=45)
for p in ax.patches: 
    ax.annotate(np.round(p.get_height(),decimals=2), 
                (p.get_x()+p.get_width()/2., p.get_height()), 
                ha='center', va='center', xytext=(0, 10), 
                textcoords='offset points')
plt.show()
```

```python
groupedv=cars.groupby('Origin').mean().reset_index()
```

```python
groupedv
```

```python
import seaborn as sns
sns.set(style="whitegrid")
ax = sns.barplot(y='Origin', x="MPG", data=groupedv)

for index, row in groupedv.iterrows():
    ax.text(row.MPG, row.name,round(row.MPG,2), color='black', ha="center")
plt.show()
```
